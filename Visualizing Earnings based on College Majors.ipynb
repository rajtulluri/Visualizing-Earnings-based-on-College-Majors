```python
import pandas as pd
import numpy as np
import matplotlib as plt
from pandas.plotting import scatter_matrix
```

```python
df = pd.read_csv('recent-grads.csv')
df
```

```python
df.dropna(inplace=True)
```

```python
df.plot(x='Sample_size',y='Median',kind='scatter',figsize=(16,8),
       title='Median vs Sample size')
df[['Sample_size','Median']].corr()
```

Between Sample size( # of full time employees) and Median of full time employee's salaries we can observe a slighlt negative correlation. As the Sample size increases the spread of Median restricts to a smaller interval. Sample size lesser than 1000 have more spread of data in terms of Median salaries.
Do Students in more popular majors make more money?
No, The negative correlation tells us that as full time employed students from a major(sample size) increase, the median salary spread is lesser.

```python
df.plot(x='Sample_size',y='Unemployment_rate',kind='scatter',figsize=(16,8),
       title='Unemployment_rate vs Sample size')
df[['Sample_size','Unemployment_rate']].corr()
```

Between Sample size and Unemployment rate, we observe there is a slightly positive correlation in data. As Sample size increases, the unemployment rates also increases slightly. We see at lower values of Sample size, the uneployment rate does have a larger spread but concentration of data is at lower values as compared to the spread of data as Sample size increases.

```python
df.plot(x='Full_time',y='Median',kind='scatter',figsize=(16,8),
       title='Median vs Full time employees')
df[['Full_time','Median']].corr()
```

Between full time employees and median salary of full time employees, we can observe - The correlation is slightly negative. As the sample size increases, the spread of data narrows down to a strict interval. Majors with lesser full time employees (lesser than 50000) have a wide spread of median salary.

```python
df.plot(x='ShareWomen',y='Unemployment_rate',kind='scatter',figsize=(16,8),
       title='Unemployment rate vs Share of Women')
df[['ShareWomen','Unemployment_rate']].corr()
```

Between Unemployment rate and Share of Women in Majors, we observe a slightly positive correlation. If we fit a straight line through the data (regression) we can see the slope is positive. As Share of women increase in the major, the unemployment rate also increases slightly.

```python
df.plot(x='Men',y='Median',kind='scatter',figsize=(16,8),
       title='Median vs Men')
df[['Men','Median']].corr()
```

Between Men graduates of a major and Median salary of full time employees, we observe a positive correlation. As Men graduates increase in number, the Median salaries also increase by a slight amount.

```python
df.plot(x='Women',y='Median',kind='scatter',figsize=(16,8),
       title='Median vs Women')
df[['Women','Median']].corr()
```

Between Women graduates and Median salaries of full time employees, we observe a negative correlation. As the number of Women increases, the Median salary spread restricts to an interval slightly lesser than the spread of Median values with Women lesser than 50000.

```python
df.plot(x='ShareWomen',y='Median',kind='scatter',figsize=(16,8),
       title='Median vs Share of Women')
df[['ShareWomen','Median']].corr()
```

Between Median salaries of full time employees and Share of women graduates, we observe a heavy negative correlation. The Median salaries clearly decreases as Share of Women increases.
Do students that majored in subjects with majority female make more money?
No, As we can see in the plot, there is a negative correlation.
